```python
g = M.metric()
```

```python
u = var('u', domain = 'real')
```

```python
v = var('v', domain = 'real')
```

```python
M = function('M')(u,v)
```

```python
U = function('U')(u,v)
```

```python
V = function('V')(u,v)
```

```python
W = function('W')(u,v)
```

```python
g[0,1] = exp(-M)
```

```python
g[3,3] = -exp(-U-V) * cosh(W)
```

```python
g[2,2] = -exp((V-U)) * cosh(W)
```

```python
g[2,3] = exp(-U) * sinh(W)
```
